```python
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
import tensorflow as ts, keras
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, LabelEncoder, StandardScaler
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df.info()
```

```python
df.head(n=10)
```

```python
df.drop(columns=['Name','Cabin','Ticket','PassengerId','SibSp','Parch','Embarked'],inplace=True)
```

```python
df.Age = df.Age.fillna(round(df.Age.mean()))
```

```python
df.head(n=10)
```

```python
inputs = df.drop(columns=['Survived'])
```

```python
inputs
```

```python
inputs.Sex = inputs.Sex.map({'male': 1, 'female': 2})
```

```python
inputs
```

```python
target = df.Survived
```

```python
target
```

```python
x_train, x_test, y_train, y_test = train_test_split(inputs,target,test_size=0.1)
```

```python
model = DecisionTreeClassifier()
model
```

```python
model.fit(x_train,y_train)
```

```python
x_test.head(n=10)
```

```python
model.predict(x_test)
```

```python
model.score(x_test,y_test)
```
